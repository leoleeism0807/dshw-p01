## е®үиЈ…еҗҺжЈҖжҹҘ

```python
import importlib

packages = [
    'akshare', 'baostock', 'pandas', 'numpy', 'matplotlib',
    'seaborn', 'statsmodels', 'scipy', 'pyarrow', 'openpyxl'
]

for pkg in packages:
    try:
        module = importlib.import_module(pkg)
        version = getattr(module, '__version__', 'unknown')
        print(f'{pkg:<12} OK  version={version}')
    except Exception as e:
        print(f'{pkg:<12} FAILED  {e}')

print('\nеҰӮжһңиҝҷйҮҢе…ЁйғЁжҳҫзӨә OKпјҢйҖҡеёёе°ұеҸҜд»Ҙз»§з»ӯиҝҗиЎҢйЎ№зӣ® Notebook дәҶгҖӮ')
print('иӢҘ Jupyter жҸҗзӨәйңҖиҰҒйҮҚеҗҜеҶ…ж ёпјҢиҜ·е…ҲйҮҚеҗҜеҶ…ж ёеҶҚз»§з»ӯгҖӮ')
```
